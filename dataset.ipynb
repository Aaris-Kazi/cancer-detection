```python
import cv2 as cv
import pandas as pd
import os
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
df = pd.read_csv('train_labels.csv')
```

```python
files = os.listdir('train/')
```

```python
f = str(files[48]).strip('.tif')
stat = df.label[df['id'] == f]
stat
# j = str(stat).split('\n')
# j = str(j[0]).split('    ')
# j
```

```python
data =[]
status =[]
file_name =[]
```

```python
for i in range(0, 15000):
    try:
        img = cv.imread("train/"+files[i])
        f = str(files[i]).strip('.tif')
        stat = df.label[df['id'] == f]
        j = str(stat).split('\n')
        j = str(j[0]).split('    ')
        status.append(j[1])
        data.append(img)
        file_name.append(f)
    except Exception:
        pass
```

```python
len(data)
```

```python
from sklearn.model_selection import train_test_split
```

```python
(x_train, x_test, y_train, y_test )= train_test_split(np.array(data), np.array(status), test_size=0.2)
```

```python
plt.bar([1,2], [len(x_train) ,len(x_test)])
plt.show
```

```python
sdata = {}
sdata['datas'] = []
sdata['status'] = []
sdata['filenames'] = []
```

```python
for i in range(len(data)):
        sdata['status'].append(status[i])
        sdata['filenames'].append(file_name[i])
        sdata['datas'].append(data[i]) 
```

```python
from joblib import dump, load
```

```python
dump(sdata, 'lymph_node_15.joblib')
```
